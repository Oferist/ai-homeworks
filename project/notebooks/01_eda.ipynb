# 01. EDA: спрос на велопрокат

Цель ноутбука - быстро проверить структуру UCI Bike Sharing Dataset, распределение целевой переменной `cnt` и основные зависимости спроса от часа, сезона, погоды и рабочих дней.

```python
from pathlib import Path
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
data = pd.read_csv(ROOT / 'data' / 'raw' / 'hour.csv')
data.head()
```

```python
data[['cnt', 'temp', 'atemp', 'hum', 'windspeed']].describe().round(3)
```

```python
hourly = data.groupby('hr', as_index=False)['cnt'].mean()
sns.lineplot(data=hourly, x='hr', y='cnt', marker='o')
plt.title('Средний спрос по часам')
plt.show()
```

```python
weather = data.groupby('weathersit', as_index=False)['cnt'].mean()
season = data.groupby('season', as_index=False)['cnt'].mean()
display(weather)
display(season)
```

Выводы: спрос заметно зависит от часа дня, температуры, рабочего дня и погоды. В обучении не используются `casual` и `registered`, потому что они являются компонентами целевой переменной `cnt` и дали бы утечку.
